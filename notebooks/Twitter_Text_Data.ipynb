# Text Cleaning

```python
tweet_text_document = tweet_data.text.tolist()
```

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
def textPrecessing(text):
    text = text.lower()
    text = " ".join([word for word in text.split(" ") if not word.startswith("http")])
    for c in string.punctuation:
        text = text.replace(c, ' ')

    wordLst = nltk.word_tokenize(text)

    filtered = [w for w in wordLst if w not in stopwords.words('english')]

    ps = PorterStemmer()
    filtered = [ps.stem(w) for w in filtered]

    return " ".join(filtered)
```

```python
output = [textPrecessing(doc) for doc in tweet_text_document]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(stop_words='english', max_features=100)
X_train_counts = count_vect.fit_transform(output)
```

# Applying LDA

```python
from sklearn.decomposition import LatentDirichletAllocation
n_topic = 5
lda = LatentDirichletAllocation(n_components=n_topic, 
                                max_iter=50,
                                learning_method='batch')
result= lda.fit_transform(X_train_counts) 
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))

n_top_words=20
tf_feature_names = count_vect.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```
